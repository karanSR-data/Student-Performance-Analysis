## 🧹 Data Cleaning

```python
# Unique vales in categorical columns
print(spa['test preparation course'].unique())
print(spa['test preparation course'].dtype)
```

```python
# Rename columns using direct functon 
spa.rename(columns={'gender':'sex'})
spa.head(5)
```

## 📊 Data Analysis

```python
# Average Score Analysis

gender_avg = spa.groupby('gender')[['math score','reading score','writing score']].mean()
gender_avg
```

```python
# Parental Education vs Student Scores

parental_level_of_education = spa.groupby('parental level of education')[['math score','reading score','writing score']].mean().sort_values
parental_level_of_education
```

```python
# Test Preparation Course Impact

test_preparation_course = spa.groupby('test preparation course')[['math score','reading score','writing score']].mean().sort_values
test_preparation_course
```

```python
# Gender-wise Score Distribution
spa.groupby('gender')[['math score']].mean()
```

```python
spa
```

## 📉 Visualizations Using Matplotlib and Seaborn

```python
sns.set_style('whitegrid')

# Bar Chart – Gender-wise Average Scores

# Creating a barplot
sns.barplot(data=gender_avg ,x='gender' ,y='math score')
plt.title('Gender wise average score')
plt.xticks(rotation=0)
plt.show()
```

```python
# Creating a histplot
sns.displot(data=spa, x='math score', kind='hist',hue='gender')
```

```python
# Boxplot – Math Score Distribution by Test Prep

sns.boxplot(data=spa,x='math score',y='test preparation course')
plt.title('Math score by Test preparation course',color='grey')
```

```python
# Histogram – Reading Score Distribution

sns.histplot(data=spa , x='reading score',bins=10)
plt.title('Distribution of Reading Score')
plt.xlabel('Reading Score')
plt.show()
```

```python
# Line Plot – Average Scores by Parental Education Level

sns.lineplot(data = spa ,y='parental level of education', x='math score')
```
